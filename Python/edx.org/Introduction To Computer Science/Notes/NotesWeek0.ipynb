**Algorithms**

There might be an application on the phones that store our contacts, with their names, numbers sorted alphabetically.<br>
The old-school equivalent might be a phone book, a printed copy of names and phone numbers.

Opening the book and start from the first page, looking for a name one page at a time.<br>
This algorithm would be correct, since we will eventually find the name if it's in the book<br>

We might flip through the book two pages at a time, but this algorythm would not be correct since we might skip the page with the name on it.<br>
Another algorithm would be opening the phone book to the middle, decide whether the name will be in the left or the right half of the book. (As the book is alphabetized)<br>
and reduce the size of the challange, by half. This strategy could be repeated unntill the name were found, dividing the problem in half each time.


**Algorithm**

1. Searching on one page at the time, Time increases lineraly as the size of the challange increases<br> n is a number representing the size of the challange, so with n pages in the book, it takes up to n steps to find the name.

2. Searching two pages at the time. The slope is less steps, but still linear. Now its roughly n/2 steps. since we flip two pages at a time.

3. Dividing the phone book in half each time. With a different relationship between the size of the challange <br> and the time to solve it. If the phone book doubled in size from 1 - 2000 pages, we would only need one more step to find the name.
