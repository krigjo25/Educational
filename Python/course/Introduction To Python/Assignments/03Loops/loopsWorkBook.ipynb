##  camelCase

In some languages, it’s common to use camel case<br>
(otherwise known as “mixed case”) for variables’<br>
names when those names comprise multiple words, <br>
whereby the first letter of the first word is<br>
lowercase but the first letter of each subsequent<br>
word is uppercase. For instance, whereas a variable<br>
for a user’s name might be called name, a variable<br>
for a user’s first name might be called firstName,<br>
and a variable for a user’s preferred first name <br>
(e.g., nickname) might be called preferredFirstName.

Python, by contrast, recommends snake case, whereby<br>
words are instead separated by underscores (_), with<br>
all letters in lowercase. For instance, those same<br>
variables would be called name, first_name, and<br>
preferred_first_name, respectively, in Python.

### The Case
In a file called camel.py, implement a program that<br>
prompts the user for the name of a variable in camel case<br>
and outputs the corresponding name in snake case. Assume<br>
that the user’s input will indeed be in camel case.

```python
#   Camel.py

#   Prompting the user with a input
prmpt = input('')

# Handling the user prompt


for i in prmpt:

    #   Converting camelcase to snakecase
    if i != i.lower():
        i = f'_{i}'.lower()
    
    print (i, end='')
```

##  Cook Machine  x?

Suppose that a machine sells bottles of Coca-Cola (Coke)<br>
for 50 cents and only accepts coins in these denominations:<br>
25 cents, 10 cents, and 5 cents.

In a file called coke.py, implement a program that prompts the<br>
user to insert a coin, one at a time, each time informing the user<br>
of the amount due. Once the user has inputted at least 50 cents, output<br>
how many cents in change the user is owed. Assume that the user will only<br>
input integers, and ignore any integer that isn’t an accepted denomination.


```python
#   Coke.py

#   Initializing variables
x = 50
c = 0
#   Creating a while loop to iterate the program
while x > 0:

    #   Prompting the use to input a cent
    print(f'amount due : {x}')   

    prmpt = int(input('Cent :'))


    if prmpt == 5 or prmpt == 10 or prmpt == 25 or prmpt == 50:
        x -= prmpt

    if x < 0:
        c -= x
        

print(f'amount Change : {c}')
```

##  Just setting up my twittr

When texting or tweeting, it’s not uncommon to shorten words to save time or space,<br>
as by omitting vowels, much like Twitter was originally called twttr. In a file called<br>
twttr.py, implement a program that prompts the user for a str of text and then outputs that<br>
same text but with all vowels (A, E, I, O, and U) omitted, whether inputted in uppercase or lowercase.

```python
#   twttr.py

#   Initializing a list
x = [
        'A','E',
        'I', 'O', 
        'U', 'Y'
    ]

#   Promting the user to input a text
prmpt = input().upper()


#   Iterating throught the string
for i in prmpt:

    #   Checking if there is a vowel in the string
    if i in x:

        #   Handling the string
        i = str(i.replace(f'{i}', ''))
        i.strip()

    #   Lowering the remaining characters
    i = i.lower()

    #   Printing the output
    print(i, end= '')
```
