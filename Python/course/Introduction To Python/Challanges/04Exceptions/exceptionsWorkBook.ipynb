## Fuel Gauge
Fuel gauge indicate often with fractions, just how much fuel there is in a tank.<br>
For instance 1/4 indicates that the tank is 25% 1/2 indicates the tank is 50% full,<br>
and 3/4 indicates that the tank is 75% full.

**In a file called `fuel.py` implement a program **

1. that prompts the user for a fraction, formated as `x/y`

2. Wherein each of `x` and `y` is an integer, and then outputs as a procentage rounded to the nearest integer

3. how much fuel in the tank. If, though 1% or less remains, output `E`instead to indicate that the tank is

4. essentially empty. And if 99% or more remains, output `F` instead to indicate that the tank is essentially full.

5. If, though, `X` or `Y` is not an integer, `X` is greater than `Y`, or `Y` is `0`, instead prompt the user again.<br>(It is not necessary for `Y` to be `4`.) Be sure to catch any exceptions like `ValueError` or `ZeroDivisionError`.

```python
def main():

    '''
        # Author : krigjo25
        # Date : 19.09-22
        
        #   Prompt the user for a fraction, formated as `x/y`
        #   Wherein each of `x` and `y` is an integer, and then outputs as a procentage rounded to the nearest integer
        #   how much fuel in the tank. If, though 1% or less remains, output `E`instead to indicate that the tank is
        #   essentially empty. And if 99% or more remains, output `F` instead to indicate that the tank is essentially full.
        #   If, though, `X` or `Y` is not an integer, `X` is greater than `Y`, or `Y` is `0`, instead prompt the user again.<br>(It is not necessary for `Y` to be `4`.) Be sure to catch any exceptions like `ValueError` or `ZeroDivisionError`.
    '''
    #   Prompting the user for a input
    while True:
        try:

            x, y = input('prompt : ').split('/')

                    #  Converting string into into
            x, y = int(x), int(y)

            #   Finding the procentage of the fraction
            x = (x / y) * 100

        except (ValueError, ZeroDivisionError) as e:
            print(e)
            continue

        else:
            #   Converting Fraction to integer

            x = round(x)

            #   Checking if the value is below 100%
            if x < 101:

                if x > 98:
                    print('F')
                elif x < 2:
                    print('E')
                else:
                    print(f'{x}%')

                #   Breaking out of the loop
                break

            else :
                continue

    return

main()
```

In the United States, dates are typically formatted in [month-day-year order](https://en.wikipedia.org/wiki/Date_and_time_notation_in_the_United_States) (MM/DD/YYYY),<br>
otherwise known as [middle-endian order](https://en.wikipedia.org/wiki/Endianness#Middle-endian), which is arguably bad design. Dates in that format can’t be easily sorted because the date’s year comes last instead of first.

Try sorting, for instance, `2/2/1800`, `3/3/1900`, and `1/1/2000` chronologically in any program (e.g., a spreadsheet). Dates in that format are also ambiguous.<br>
[Harvard was founded on September 8, 1636](https://www.harvard.edu/about/history/), but 9/8/1636 could also be interpreted as August 9, 1636!

Fortunately, computers tend to use [ISO 8601](https://en.wikipedia.org/wiki/ISO_8601), an international standard that prescribes that dates should be formatted in year-month-day (YYYY-MM-DD) order, no matter the country,<br> 
formatting years with four digits, months with two digits, and days with two digits, “padding” each with leading zeroes as needed.
